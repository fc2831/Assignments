```python
def kelvin_to_celsius(k):
    return k - 273.15

def celsius_to_kelvin(c):
    return c + 273.15
```

```python
kelvin_to_celsius(273.15)
```

```python
celsius_to_kelvin(0)
```

```python
def to_fahrenheit(temp, input_unit="c"):
    """
    Convert a temperature to Fahrenheit.
    input_unit can be "c" (Celsius) or "k" (Kelvin).
    """
    if input_unit == "k":
        c = kelvin_to_celsius(temp)   # use your previous function
    elif input_unit == "c":
        c = temp
    else:
        raise ValueError("input_unit must be 'c' or 'k'")

    f = (c * 9/5) + 32
    return f
```

```python
to_fahrenheit(0, input_unit="c")
```

```python
def from_fahrenheit(f, output_unit="c"):
    """
    Convert Fahrenheit to Celsius or Kelvin.
    output_unit can be "c" (Celsius) or "k" (Kelvin).
    """
    c = (f - 32) * 5/9  

    if output_unit == "c":
        return c
    elif output_unit == "k":
        return celsius_to_kelvin(c)  
    else:
        raise ValueError("output_unit must be 'c' or 'k'")
```

```python
from_fahrenheit(32, output_unit="k")  
```

```python
def ft_in_to_m(feet, inches):
    """
    Convert height in feet + inches to meters.
    """
    total_inches = feet * 12 + inches
    meters = total_inches * 0.0254
    return meters
```

```python
ft_in_to_m(6, 0)
```

```python
def m_to_ft_in(meters):
    """
    Convert meters to (feet, inches).
    Returns: (feet, inches)
    """
    total_inches = meters / 0.0254
    feet = int(total_inches // 12)  
    inches = total_inches % 12      
    return feet, inches
```

```python
m_to_ft_in(1.8288)
```

```python
test_meters = [1.69, 1.80, 2.00]

for m in test_meters:
    ft, inch = m_to_ft_in(m)
    m_back = ft_in_to_m(ft, inch)
    diff = abs(m - m_back)

    print("start:", m, "m")
    print("->", ft, "ft", inch, "in")
    print("back:", m_back, "m")
    print("difference:", diff)
    print("-----")
```
